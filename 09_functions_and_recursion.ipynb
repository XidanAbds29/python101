## 2. Default Arguments (Purbo-nirdharito maan)

Function er parameter e amra default value dite pari. Jodi keu value na dey, tokhon default ta use hobe.

````python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Karim"))               # Default greeting use hobe
print(greet("Karim", "Salam"))       # Custom greeting
````

### Bakhya (Explanation)
- `greeting="Hello"` mane jodi keu greeting na dey, tahole `"Hello"` use hobe.
- Kintu jodi custom value dey, seta use hobe.

## 3. Keyword Arguments

Amra function call korar somoy parameter er naam diye value pathate pari. Etokhon position diye pathatam.

````python
def student_info(name, age, department):
    print(f"Naam: {name}, Boyosh: {age}, Dept: {department}")

# Positional arguments (kramomanusare)
student_info("Rahim", 20, "CSE")

# Keyword arguments (naam diye - jeko kono krome)
student_info(department="EEE", name="Karim", age=21)
````

## 4. `*args` ebong `**kwargs`

Jokhon tumi jano na koyti argument ashbe, tokhon `*args` (tuple) ba `**kwargs` (dictionary) use koro.

````python
# *args: Jekono shonkhok positional argument nite pare
def total(*numbers):
    jogfol = 0
    for num in numbers:
        jogfol += num
    return jogfol

print(f"Jogfol (3 ta): {total(10, 20, 30)}")
print(f"Jogfol (5 ta): {total(1, 2, 3, 4, 5)}")
````

````python
# **kwargs: Jekono shonkhok keyword argument nite pare
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print("=== Student 1 ===")
print_info(name="Rahim", age=20, dept="CSE")
print("\n=== Student 2 ===")
print_info(name="Fatima", age=22, dept="EEE", cgpa=3.8)
````

## 5. Lambda Function (Chhoto Namhin Function)

Lambda holo one-line er anonymous function. Chhoto kaj er jonno use kora hoy.

````python
# Normal function
def square(x):
    return x * x

# Seta Lambda diye
square_lambda = lambda x: x * x

print(f"Normal function: {square(5)}")
print(f"Lambda function: {square_lambda(5)}")

# Lambda with sorting
students = [("Rahim", 85), ("Karim", 92), ("Fatima", 78)]
students.sort(key=lambda x: x[1])  # Marks diye sort
print(f"\nMarks diye sorted: {students}")
````

# Recursion (রিকার্সন)

## Recursion ki?

Recursion holo jokhon ekta function **nijeke nije call kore**. Eta ekta powerful technique ja onek complex somossa solve korte pare.

**Baastaab jibon er udahoron:**
- Tumi ekta ayna er samne daralo, ebong pechone arekta ayna ache. Tumi ayna te ayna dekhte pacho — eta recursion!
- Russian Matryoshka doll — ekta doll er bhitore arekta choto doll, tar bhitore aro choto...

**Recursion er 2 ta guruttopurno jinish:**
1. **Base Case**: Jokhon function ar nijeke call kore na — STOP!
2. **Recursive Case**: Function nijeke call kore choto somossa diye.

## 6. Prothom Recursion: Countdown

Amra 5 theke shuru kore 1 porjonto countdown korbo.

````python
def countdown(n):
    if n <= 0:         # Base case: Jokhon 0 ba tar niche
        print("BLAST OFF! 🚀")
        return
    print(n)
    countdown(n - 1)   # Recursive call: choto somossa

countdown(5)
````

### Bakhya (Explanation)
- `countdown(5)` call kore → 5 print kore → `countdown(4)` call kore
- `countdown(4)` call kore → 4 print kore → `countdown(3)` call kore
- ... emonbhabe choltei thake
- `countdown(0)` → Base case hit! "BLAST OFF!" print kore ebong return kore.

**⚠️ Shorokotha:** Base case na dile function infinitely call hote thakbe ebong Python error dibe (`RecursionError`)!

## 7. Factorial (n!) - Recursion diye

Factorial mane: `n! = n × (n-1) × (n-2) × ... × 1`

Udahoron: `5! = 5 × 4 × 3 × 2 × 1 = 120`

Recursive formula: `n! = n × (n-1)!` ebong `0! = 1` (base case)

````python
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    # Recursive case
    return n * factorial(n - 1)

# Test
for i in range(8):
    print(f"{i}! = {factorial(i)}")
````

### Bakhya (Explanation)
Dhoro amra `factorial(4)` call korlam:
```
factorial(4) = 4 × factorial(3)
             = 4 × 3 × factorial(2)
             = 4 × 3 × 2 × factorial(1)
             = 4 × 3 × 2 × 1    ← base case!
             = 24
```

## 8. Fibonacci Series - Recursion diye

Fibonacci series: `0, 1, 1, 2, 3, 5, 8, 13, 21, ...`

Formula: `fib(n) = fib(n-1) + fib(n-2)`, Base: `fib(0)=0, fib(1)=1`

````python
def fibonacci(n):
    # Base cases
    if n == 0:
        return 0
    if n == 1:
        return 1
    # Recursive case
    return fibonacci(n - 1) + fibonacci(n - 2)

# Prothom 10 ta Fibonacci number
print("Fibonacci Series:")
for i in range(10):
    print(f"fib({i}) = {fibonacci(i)}")
````

## 9. Power Function - Recursion diye

`a^b = a × a × ... (b bar)`. Recursive formula: `power(a, b) = a × power(a, b-1)`, Base: `power(a, 0) = 1`

````python
def power(base, exp):
    # Base case: jekono shonkhar ghat 0 = 1
    if exp == 0:
        return 1
    # Recursive case
    return base * power(base, exp - 1)

# Test
test_cases = [(2, 5), (3, 4), (5, 3), (10, 0)]
for b, e in test_cases:
    print(f"{b}^{e} = {power(b, e)}")
````

## 10. Sum of Digits - Recursion diye

Ekta number er digit gulo jog kora. Udahoron: `1234 → 1+2+3+4 = 10`

````python
def sum_of_digits(n):
    # Base case: Ekta digit thakle seta return koro
    if n < 10:
        return n
    # Recursive case: shesh digit + baki digit gular sum
    return (n % 10) + sum_of_digits(n // 10)

# Test
numbers = [1234, 9876, 555, 100]
for num in numbers:
    print(f"Digits of {num} er jogfol = {sum_of_digits(num)}")
````

### Bakhya (Explanation)
- `n % 10` diye shesh digit ta pai (1234 % 10 = 4)
- `n // 10` diye baki ta pai (1234 // 10 = 123)
- Ebar abar `sum_of_digits(123)` call kori — choto somossa!

## 11. GCD - Recursion diye (Euclidean Algorithm)

GCD (Greatest Common Divisor) ber korar shobcheye efficient upay holo Euclidean Algorithm.

````python
def gcd(a, b):
    # Base case: jodi b 0 hoy, tahole a holo GCD
    if b == 0:
        return a
    # Recursive case
    return gcd(b, a % b)

# Test
pairs = [(48, 18), (100, 75), (12, 8), (17, 5)]
for a, b in pairs:
    print(f"GCD({a}, {b}) = {gcd(a, b)}")
````

## Recursion vs Loop: Kothay ki use korbo?

| Feature | Loop | Recursion |
|---------|------|-----------|
| Bojha shohoj? | Haan | Na (sometimes) |
| Memory use | Kom | Beshi (stack) |
| Code length | Lamba | Chhoto |
| Best for | Simple repetition | Tree/Divide-and-conquer |
| Speed | Beshi fast | Slower (call overhead) |

**Tip:** Jodi somossa ta naturally choto choto sub-problems e bhag kora jay (jemon factorial, fibonacci, tree traversal), tahole recursion use koro. Nahole loop use koro.
