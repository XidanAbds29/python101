# Loops (Bar bar kaj kora)

Computer ek e kaj bar bar korte khub bhalo pare. Etake "looping" bola hoy.
Python e du dhoroner loop ache: `while` loop ebong `for` loop.

## 1. `while` loop

`while` loop totokhon chole jotokhon condition ta sotti (True) thake.

````python
count = 1 #initializing the variable
#("gonona holo )
while count <= 5: #condition check
    print(f"Gonona holo: {count}")
    count = count + 1#incrementing the variable

print("Gonona sesh")
````

### Bakhya (Explanation)
- Amra `count` shuru korlam 1 theke.
- Loop check korche `count <= 5` kina. Eta sotti, tai eta print korlo.
- **Guruttopurno Dhap**: `count = count + 1` count er man baralo. Eta na dile loop ti sarajibon cholte thakto!

## 2. `for` loop

`for` loop use kora hoy kono sequence ba list er upore iterate (ekta ekta kore jawa) korar jonno. Amra prayi `range()` er sathe eta use kori.

````python
# range(5) amader 0, 1, 2, 3, 4 number gulo dey
for i in range(5):
    print(f"Eta holo loop number {i}")

    
````

### Bakhya (Explanation)
- `range(5)` shoonyo (0) theke shuru kore 5 er ag porjonto number dey.
- `i` holo ekta variable ja protibar ekta kore notun number ney sequence theke.

## 4. `break` and `continue`

- `break`: Loop ta ekhoni bondho koro.
- `continue`: Loop er baki ongsho skip kore porer bare chole jao.

````python
for i in range(10):
    if i == 5:
        break  # Verify: Jokhon i 5 hobe, tokhon loop bondho hoye jabe
    print(i)
````

````python
for i in range(5):
    if i == 2:
        continue # Verify: Eta 2 ke skip korbe, kintu 3 ar 4 print korbe
    print(i)
````

````python
## Extra Problem: Armstrong Numbers

### Problem Statement (সমস্যার বিবৃতি)

**Question:** Write a program to print out all Armstrong numbers between 1 and 10000. If the sum of cubes of each digit of the number is equal to the number itself, then the number is called an Armstrong number. For example, 153 = (1³) + (5³) + (3³).

**বাংলা ব্যাখ্যা:**
- Armstrong number একটি সংখ্যা যেখানে প্রতিটি ডিজিটের cubes এর যোগফল সংখ্যাটির সমান
- উদাহরণ: 153 = 1³ + 5³ + 3³ = 1 + 125 + 27 = 153
- তিন ডিজিট Armstrong numbers: 153, 370, 371, 407
````

````python
# Solution: Find all Armstrong Numbers between 1 and 10000
print("=== Armstrong Numbers বের করা (1 থেকে 10000) ===\n")

armstrong_numbers = []

for num in range(1, 10001):
    # Convert number to string to iterate through digits
    num_str = str(num)
    num_digits = len(num_str)
    
    # Calculate sum of each digit raised to power of number of digits
    sum_of_powers = 0
    for digit in num_str:
        digit_int = int(digit)
        sum_of_powers += digit_int ** num_digits
    
    # Check if it's an Armstrong number
    if sum_of_powers == num:
        armstrong_numbers.append(num)

print(f"Armstrong numbers between 1 and 10000:")
print(armstrong_numbers)
print(f"\nTotal Armstrong numbers found: {len(armstrong_numbers)}")

# Show details for each Armstrong number
print("\nDetailed breakdown:")
for num in armstrong_numbers:
    num_str = str(num)
    num_digits = len(num_str)
    calculation = " + ".join([f"{d}^{num_digits}" for d in num_str])
    powers = " + ".join([str(int(d)**num_digits) for d in num_str])
    print(f"{num}: {calculation} = {powers} = {num}")
````

## Pattern Printing Problems (প্যাটার্ন প্রিন্টিং সমস্যা)

### Problem: Different Triangle and Rectangle Patterns

**Question:** Write a program to show the following triangle/rectangle of '*'s or numbers. Take n as input from the user to determine the number of rows of the structure.

**বাংলা:** বিভিন্ন প্যাটার্ন প্রিন্ট করার প্রোগ্রাম লিখুন। ব্যবহারকারী থেকে n ইনপুট নিয়ে বিভিন্ন আকৃতির প্যাটার্ন তৈরি করুন।

#### Pattern 1: Ascending Triangle with Stars (ঊর্ধ্বমুখী তারকা ত্রিভুজ)

```
*
* * *
* * * * *
* * * * * * *
* * * * * * * * *
```

প্রতিটি সারিতে (2*i - 1) টি তারকা আছে যেখানে i হল সারির নম্বর।

````python
# Solution - Pattern 1: Ascending Triangle with Stars
print("=== Pattern 1: Ascending Triangle with Stars ===")
n = int(input("Enter n (number of rows): "))

for i in range(1, n + 1):
    for j in range(2 * i - 1):
        print("*", end=" ")
    print()
````

#### Pattern 2: Filled Rectangle with Stars (পূর্ণ আয়তক্ষেত্র)

```
* * * * *
* * * * *
* * * * *
* * * * *
* * * * *
```

n×n আয়তক্ষেত্র তৈরি করা।

````python
# Solution - Pattern 2: Filled Rectangle with Stars
print("\n=== Pattern 2: Filled Rectangle with Stars ===")
n = int(input("Enter n (number of rows/columns): "))

for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()
````

#### Pattern 3: Hollow Rectangle with Stars (ফাঁকা আয়তক্ষেত্র)

```
* * * * *
* * * * *
* * * * *
* * * * *
* * * * *

অথবা

* * * * *
*       *
*       *
*       *
* * * * *
```

শুধুমাত্র প্রান্তে তারকা প্রিন্ট করা হয়।

````python
# Solution - Pattern 3: Hollow Rectangle with Stars
print("\n=== Pattern 3: Hollow Rectangle with Stars ===")
n = int(input("Enter n (number of rows/columns): "))

for i in range(n):
    for j in range(n):
        # Print stars only on borders (first/last row or first/last column)
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

#### Pattern 4: Number Pyramid (সংখ্যার পিরামিড)

```
1
1 2 1
1 2 3 2 1
1 2 3 4 3 2 1
1 2 3 4 5 4 3 2 1
```

প্রতিটি সারিতে 1 থেকে i পর্যন্ত সংখ্যা এবং তারপর আবার i-1 থেকে 1 পর্যন্ত।

````python
# Solution - Pattern 4: Number Pyramid
print("\n=== Pattern 4: Number Pyramid ===")
n = int(input("Enter n (number of rows): "))

for i in range(1, n + 1):
    # Ascending numbers
    for j in range(1, i + 1):
        print(j, end=" ")
    # Descending numbers (excluding the peak)
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()
````

#### Pattern 5: Descending Triangle (অবরোহী ত্রিভুজ)

```
* * * * * * * * *
* * * * * * *
* * * * *
* * *
*
```

প্রথম সারিতে সবচেয়ে বেশি তারকা, শেষে শুধুমাত্র একটি।

````python
# Solution - Pattern 5: Descending Triangle
print("\n=== Pattern 5: Descending Triangle ===")
n = int(input("Enter n (number of rows): "))

for i in range(n, 0, -1):
    for j in range(2 * i - 1):
        print("*", end=" ")
    print()
````

#### Pattern 6: Symmetric Number Diamond (ডায়মন্ড আকৃতি)

```
1
1 2 1
1 2 3 2 1
1 2 3 4 3 2 1
1 2 3 5 4 3 2 1
1 2 3 4 3 2 1
1 2 3 2 1
1 2 1
1
```

উপরের অর্ধেক ascending, নিচের অর্ধেক descending number pyramid।

````python
# Solution - Pattern 6: Symmetric Number Diamond
print("\n=== Pattern 6: Symmetric Number Diamond ===")
n = int(input("Enter n (number of rows in half): "))

# Ascending part
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()

# Descending part
for i in range(n - 1, 0, -1):
    for j in range(1, i + 1):
        print(j, end=" ")
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()
````
