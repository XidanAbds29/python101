# Loops (Bar bar kaj kora)

Computer ek e kaj bar bar korte khub bhalo pare. Etake "looping" bola hoy.
Python e du dhoroner loop ache: `while` loop ebong `for` loop.

## 1. `while` loop

`while` loop totokhon chole jotokhon condition ta sotti (True) thake.

```python
count = 1

while count <= 5:
    print(f"Gonona holo: {count}")
    count = count + 1
```

### Bakhya (Explanation)
- Amra `count` shuru korlam 1 theke.
- Loop check korche `count <= 5` kina. Eta sotti, tai eta print korlo.
- **Guruttopurno Dhap**: `count = count + 1` count er man baralo. Eta na dile loop ti sarajibon cholte thakto!

## 2. `for` loop

`for` loop use kora hoy kono sequence ba list er upore iterate (ekta ekta kore jawa) korar jonno. Amra prayi `range()` er sathe eta use kori.

```python
# range(5) amader 0, 1, 2, 3, 4 number gulo dey
for i in range(5):
    print(f"Eta holo loop number {i}")

    
```

### Bakhya (Explanation)
- `range(5)` shoonyo (0) theke shuru kore 5 er ag porjonto number dey.
- `i` holo ekta variable ja protibar ekta kore notun number ney sequence theke.

## 3. Looping through a list (List er upore loop)

Amra `for` loop use kore list er item gulo access korte pari.

```python
fruits = ["Apple", "Banana", "Cherry"]

for fruit in fruits:
    print("Ami %s khete pochondo kori" % fruit)
```

## 4. `break` and `continue`

- `break`: Loop ta ekhoni bondho koro.
- `continue`: Loop er baki ongsho skip kore porer bare chole jao.

```python
for i in range(10):
    if i == 5:
        break  # Verify: Jokhon i 5 hobe, tokhon loop bondho hoye jabe
    print(i)
```

```python
for i in range(5):
    if i == 2:
        continue # Verify: Eta 2 ke skip korbe, kintu 3 ar 4 print korbe
    print(i)
```

```python
#gcd,lcm of three numbers 
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
c=int(input("Enter third number: "))

gcd=1
for i in range(1, min(a,b,c)+1):
    if a%i==0 and b%i==0 and c%i==0:
        gcd=i
        final_gcd=gcd
lcm=(a*b*c)//final_gcd
#gcd*lcm=a*b*c
print(f"GCD: {final_gcd}, LCM: {lcm}")

```

```python
#Write a program to determine all prime numbers within the range [a-b], where a & b are input 
#through the keyboard.  

a=int(input("Enter the starting number (a): "))
b=int(input("Enter the ending number (b): "))
for num in range(a, b+1):
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if (num % i) == 0:
                break
        else:
            print(f"{num} is a prime number")
```

```python
#Write a program to find out the first n perfect number, where n is the input from the user. 

n=int(input("Enter the number of perfect numbers to find: "))
count = 0
num = 2

while count < n:
    sum_of_divisors = 0
    for i in range(1, num):
        if num % i == 0:
            sum_of_divisors += i
    if sum_of_divisors == num:
        print(f"Perfect number {count + 1}: {num}")
        count += 1
    num += 1   
```
