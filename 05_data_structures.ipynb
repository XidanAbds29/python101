# Data Structures: Lists and Dictionaries

Amader onek somoy eksathe onek gulo data rakhte hoy. List ar Dictionary holo best upay eta korar jonno.

## 1. Lists (Talika)

List holo items er ekta shajano collection.

```python
my_list = [10, 20, 30, 40]
```

### Item khuje pawa (Indexing)
Amra position diye item khuje pete pari. **Mone rakhben: Python 0 theke gonona shuru kore!**

```python
print(my_list[0])  # Prothom item ti print korbe
print(my_list[2])  # Tritiyo item ti print korbe
```

### Notun Item Jog Kora
List er sheshe notun item add korar jonno `.append()` use kora hoy.

```python
my_list.append(50)
print(my_list)
```

## 2. Dictionaries (Shobdakos)

Dictionary te data "Key-Value" pair hisebe thake. Eta ashol dictionary r motoi: apni ekta word (key) khujen tar mane (value) janar jonno.

```python
student = {
    "name": "John",
    "age": 25,
    "course": "Computer Science"
}
```

### Accessing Values
Amra key use kore value pete pari.

```python
print(student["name"])
print(student["course"])
```

### Bakhya (Explanation)
- `student["name"]` dictionary te `"name"` key ti khuje ebong tar value `"John"` ferot dey.
- Jodi emon kono key chan ja nei, tokhon Python error dibe.
