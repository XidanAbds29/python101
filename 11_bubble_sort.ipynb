# Bubble Sort (বাবল সর্ট)

## Bangla Explanation (বাংলা ব্যাখ্যা)

**Sorting** mane data ke choto theke boro (ba boro theke choto) krome sajano. **Bubble Sort** holo shobcheye sadharon sorting algorithm.

**Baastaab jibon er udahoron:**
- Dhoro tomar kache 5 ta card ache random krome. Tumi pashapashi duitake compare korcho — jodi baam er ta boro hoy, tahole swap korcho. Bar bar korte korte shobcheye boro ta sheshe chole jay — jemon pani te bubble upore uthe jay!

## 1. Bubble Sort er Niyom (Rules)

1. Array er **prothom** theke shuru koro.
2. **Pashapashi duti item** compare koro.
3. Jodi **baam er ta boro** hoy, tahole **swap** (odik-badol) koro.
4. Ebar **porer pair** e jao.
5. Puro array te ekbar gele, **shobcheye boro item ta sheshe** chole jay.
6. Abar prothom theke koro (kintu shesher item ta abar check korte hobe na).
7. Jotokhon na kono swap er dorkar na pore, totokhon chalao.

**Naam keno Bubble Sort?**
- Jemon pani te bubble upore uthe jay, temon e boro item gulo dheere dheere shesher dike chole jay!

## 2. Basic Bubble Sort

```python
def bubble_sort(arr):
    """
    Bubble Sort: Pashapashi compare kore sajano
    """
    n = len(arr)
    
    for i in range(n - 1):           # n-1 bar pass korte hobe
        for j in range(n - 1 - i):   # Protita pass e shesh er i ta skip
            if arr[j] > arr[j + 1]:  # Baam er ta boro hole swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
    return arr

# Test
numbers = [64, 34, 25, 12, 22, 11, 90]
print(f"Sort er aage: {numbers}")

sorted_numbers = bubble_sort(numbers.copy())
print(f"Sort er pore: {sorted_numbers}")
```

### Bakhya (Explanation)
- **Outer loop** (`i`): Kotogulo pass hobe ta niyontron kore. n-1 ta pass laage.
- **Inner loop** (`j`): Protita pass e pashapashi pair gulo compare kore.
- `n - 1 - i`: Protita pass e shesh diker i ta item already sorted, tai eder skip kori.
- `arr[j], arr[j+1] = arr[j+1], arr[j]`: Python er swap trick — temp variable lagbe na!

## 3. Step by Step Visualization

```python
def bubble_sort_verbose(arr):
    """Protita step dekhano version"""
    n = len(arr)
    arr = arr.copy()
    total_swaps = 0
    
    print(f"Shuru: {arr}\n")
    
    for i in range(n - 1):
        print(f"═══ Pass {i + 1} ═══")
        swapped = False
        
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                print(f"  Compare arr[{j}]={arr[j]} > arr[{j+1}]={arr[j+1]} → SWAP! 🔄")
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
                total_swaps += 1
            else:
                print(f"  Compare arr[{j}]={arr[j]} ≤ arr[{j+1}]={arr[j+1]} → thik ache ✓")
        
        print(f"  Pass {i+1} er por: {arr}")
        print(f"  (Sorted element: {arr[n-1-i]} position e boshe geche)\n")
        
        if not swapped:
            print("  Kono swap hoyni — Array already sorted! 🎉")
            break
    
    print(f"Final Sorted Array: {arr}")
    print(f"Total Swaps: {total_swaps}")
    return arr

# Test (choto array diye)
arr = [5, 3, 8, 1, 2]
bubble_sort_verbose(arr)
```

## 4. Optimized Bubble Sort

Jodi kono pass e **ekta o swap na hoy**, tahole array already sorted! Tokhon aage theke bondho kore dite pari.

```python
def bubble_sort_optimized(arr):
    """
    Optimized Bubble Sort: Jodi sorted hoye jay, tahole aage theke bondho kore dey
    """
    n = len(arr)
    arr = arr.copy()
    
    for i in range(n - 1):
        swapped = False       # Flag: kono swap hoyeche ki na
        
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        
        # Jodi kono swap na hoy, array sorted!
        if not swapped:
            print(f"Pass {i+1} e kono swap hoyni — early exit!")
            break
    
    return arr

# Test 1: Already nearly sorted array
nearly_sorted = [1, 2, 3, 5, 4]
print(f"Nearly sorted: {nearly_sorted}")
result = bubble_sort_optimized(nearly_sorted)
print(f"Result: {result}\n")

# Test 2: Already sorted array
sorted_arr = [1, 2, 3, 4, 5]
print(f"Already sorted: {sorted_arr}")
result2 = bubble_sort_optimized(sorted_arr)
print(f"Result: {result2}")
```

### Bakhya (Explanation)
- `swapped = False` protita pass er shuru te set kori.
- Jodi ekta o swap hoy, tahole `swapped = True` hoy.
- Pass sheshe jodi `swapped` still `False` thake, mane array already sorted — loop bondho!
- Eta **best case** te O(n) time e kaj kore (already sorted array).

## 5. Descending Order e Sort (Boro theke Choto)

```python
def bubble_sort_descending(arr):
    """Boro theke choto krome sajano"""
    n = len(arr)
    arr = arr.copy()
    
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] < arr[j + 1]:  # Choto ta baam e gele swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
    return arr

numbers = [34, 12, 78, 45, 90, 23]
print(f"Original: {numbers}")
print(f"Ascending (choto → boro): {bubble_sort(numbers.copy())}")
print(f"Descending (boro → choto): {bubble_sort_descending(numbers)}")
```

## 6. String Array Sort

```python
# Bubble sort strings er upore o kaj kore!
names = ["Rahim", "Anika", "Zahir", "Fatima", "Bilal"]

n = len(names)
sorted_names = names.copy()

for i in range(n - 1):
    for j in range(n - 1 - i):
        if sorted_names[j] > sorted_names[j + 1]:
            sorted_names[j], sorted_names[j + 1] = sorted_names[j + 1], sorted_names[j]

print(f"Original: {names}")
print(f"Sorted:   {sorted_names}")
```

## 7. Time Complexity (সময় জটিলতা)

| Case | Complexity | Kokhon hoy? |
|------|-----------|-------------|
| **Best** | O(n) | Array already sorted (optimized version) |
| **Average** | O(n²) | Random krome thakle |
| **Worst** | O(n²) | Ulta krome sorted thakle |
| **Space** | O(1) | Extra kono array lagbe na |

**n² keno?** Dhoro n = 1000 hole:
- Outer loop: 999 bar
- Inner loop: average ~500 bar
- Total: ~499,500 ta comparison!

**Tai Bubble Sort boro data er jonno slow**, kintu **shikhar jonno khub bhalo** ebong **choto data te thik ache**.

## Practice Problems (অনুশীলন সমস্যা)

### Problem 1: Student marks sort kore top 3 ber koro

**Question (প্রশ্ন):** Students der marks ache ekta array te. Bubble sort kore top 3 marks dekhao.

```python
# Solution
marks = [78, 92, 45, 88, 95, 67, 73, 99, 82, 61]
print(f"Shob marks: {marks}")

# Bubble sort (descending)
sorted_marks = marks.copy()
n = len(sorted_marks)
for i in range(n - 1):
    for j in range(n - 1 - i):
        if sorted_marks[j] < sorted_marks[j + 1]:
            sorted_marks[j], sorted_marks[j + 1] = sorted_marks[j + 1], sorted_marks[j]

print(f"Sorted (boro → choto): {sorted_marks}")
print(f"\n🏆 Top 3 marks:")
for i in range(3):
    print(f"  {i+1}st: {sorted_marks[i]}")
```

### Problem 2: Count total swaps

**Question (প্রশ্ন):** Ekta array ke bubble sort korte koyti swap lagbe ta count koro.

```python
# Solution
def count_swaps(arr):
    arr = arr.copy()
    n = len(arr)
    swaps = 0
    
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    
    return arr, swaps

# Test different arrays
test_arrays = [
    [5, 4, 3, 2, 1],   # Worst case (ulta sorted)
    [1, 2, 3, 4, 5],   # Best case (already sorted)
    [3, 1, 4, 1, 5],   # Random
]

for arr in test_arrays:
    sorted_arr, swap_count = count_swaps(arr)
    print(f"{arr} → {sorted_arr} | Swaps: {swap_count}")
```

### Problem 3: Sort kore median ber koro

**Question (প্রশ্ন):** Ekta number array ke bubble sort kore tar median (moddho-maan) ber koro.

```python
# Solution
def find_median(arr):
    # Bubble sort
    arr = arr.copy()
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
    # Median ber kora
    if n % 2 == 1:
        # Bijor shonkha hole majher ta
        median = arr[n // 2]
    else:
        # Jor shonkha hole majher dutar average
        median = (arr[n // 2 - 1] + arr[n // 2]) / 2
    
    return arr, median

# Test
arr1 = [7, 3, 1, 9, 5]
sorted1, med1 = find_median(arr1)
print(f"Array: {arr1} → Sorted: {sorted1} → Median: {med1}")

arr2 = [8, 2, 4, 6]
sorted2, med2 = find_median(arr2)
print(f"Array: {arr2} → Sorted: {sorted2} → Median: {med2}")
```
