# Conditional Logic (Shorto)

Ei notebook e amra shikbo kivabe `if`, `elif`, ebong `else` use kore code er moddhe siddhanto (decision) nite hoy.

## 1. `if` statement

Amra `if` tokhoni use kori jokhon amra chai jeno code ti shudhu matro condition sotti (True) holei chole.

```python
age = 20

if age >= 18:
    print("Apni ekjon prapto boyosko.")
```

### Bakhya (Explanation)
- `age >= 18` check korche boyosh ki 18 er soman ba beshi kina.
- Notice korun `print(...)` line ti ektu dan dike sorano (**indented**). Er mane holo eta `if` er under-e ache.
- Jodi `age` 10 hoto, tahole kichui print hoto na.

## 2. `else`

`if` er condition jodi false (mithye) hoy, tokhon amra `else` use kori onno kichu korar jonno.

```python
age = 15

if age >= 18:
    print("Apni ekjon prapto boyosko.")
else:
    print("Apni ekhono prapto boyosko hon ni.")
```

### Bakhya (Explanation)
- Jodi `age` 18 ba tar beshi hoy, prothom line ti chapabe.
- Na hole (else), porer line ti chapabe.

## 3. Comparison Operators (Tulona korar chinho)

Duto jinisher moddhe tulona korar jonno amra egulo use kori:

- `==` : Soman kina (Kheyal rakhben ekhane duto soman chinho!)
- `!=` : Soman NOY kina
- `>` : Boro kina
- `<` : Choto kina
- `>=` : Boro ba soman kina
- `<=` : Choto ba soman kina

```python
x = 10
y = 20 

if x == 10:
    print("x er man 10")

if x != y:
    print("x ebong y soman noy")

if x==10 and x!=y:
    print("x is 10 and x is not equal to y")    
```

## 4. Logical Operators: `and`, `or`

Amra eksathe onek gulo condition check korte pari.

```python
score = 85
attendance = 90

if score > 80 and attendance > 80:
    print("Apni distinction peyechen!")

if score > 80 or attendance > 80:
    print("Apni distinction peyechen!")
```

### Bakhya (Explanation)
- `and`: DUTO condition e sotti hote hobe.
- `or`: Jekono EKTA condition sotti holei hobe.
- `not`: Eta condition take ultiye dey (True ke False banay).

```python
x=int(input("enter a number: "))

if x % 2 == 0:
    print(f"{x} is an even number.")
else:
    print(f"{x} is an odd number.")
```

```python
cost_price=float(input("Enter cost price: "))
selling_price=float(input("Enter selling price: "))

if selling_price > cost_price:
    profit = selling_price - cost_price
    print(f"Profit: {profit}")
elif selling_price < cost_price:
    loss = cost_price - selling_price
    print(f"Loss: {loss}")
else:
    print("No profit, no loss")
```

```python
#According to the Gregorian calendar, it was Monday on the date 01/01/1900. If any year is input 
#through the keyboard, write a program to find out what is the day on 1st January of that year.

year=int(input("Enter a year: "))

days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]


total_days = 0
for y in range(1900, year):
    if (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0):
        total_days += 366
    else:
        total_days += 365

day_index = total_days % 7
print(f"1st January {year} is a {days_of_week[day_index]}")
```
