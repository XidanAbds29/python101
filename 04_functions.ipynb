# Functions (Kaj er block)

Functions holo punoray byoboharjoggo (reusable) code er block. Bar bar aki code na likhe amra seta function er bhitore rakhi ebong dorkar porle call kori.

## 1. Function Banano (Defining)

Function bananor jonno amra `def` keyword use kori.

```python
def say_hello():
    print("Hello! Course e swagotom.")
```

### Bakhya (Explanation)
- `def say_hello():`: Amra `say_hello` name ekta function banalam.
- Bhitorer code obosshoi **indented** (dan dike sorano) hote hobe.
- Ei block ta run korle kintu kichu print hobe na! Amra shudhu function ta *define* korlam.

## 2. Function ke daka (Calling)

Function use korar jonno amader setake "call" korte hoy.

```python
say_hello()
say_hello()
```

### Bakhya (Explanation)
- Amra function er nam likhi ebong tarpor bracket `()` dei.
- Amra duibar call korechi tai eta duibar print korbe.

## 3. Arguments (Tottho pathano)

Amra function er bhitore tottho (data) pathate pari jate kore function ti alada alada kaj korte pare.

```python
def greet_person(name):
    print("Hello, %s!" % name)
```

```python
greet_person("Alice")
greet_person("Bob")
```

### Bakhya (Explanation)
- Amra definition er somoy bracket er bhitore `name` likhechi.
- Jokhon amra `greet_person("Alice")` call kori, tokhon `"Alice"` value ti function er bhitore `name` variable e chole jay.

## 4. Return Values (Folafol ferot deya)

Function prayi kichu hiseb-nikesh kore ebong folafol amader ferot dey. Er jonno amra `return` use kori.

```python
def add_numbers(x, y):
    result = x + y
    return result
```

```python
suma = add_numbers(5, 3)
print("Jogfol holo: %d" % suma)
```

### Bakhya (Explanation)
- `return result` value tike ferot pathay jekhan theke function call kora hoyechilo.
- `suma = add_numbers(5, 3)` er mane `suma` variable er man hobe 8.
