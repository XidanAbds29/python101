# Functions (Kaj er block)

Functions holo punoray byoboharjoggo (reusable) code er block. Bar bar aki code na likhe amra seta function er bhitore rakhi ebong dorkar porle call kori.

## 1. Function Banano (Defining)

Function bananor jonno amra `def` keyword use kori.

```python
def say_hello():
    print("Hello! Course e swagotom.")
```

### Bakhya (Explanation)
- `def say_hello():`: Amra `say_hello` name ekta function banalam.
- Bhitorer code obosshoi **indented** (dan dike sorano) hote hobe.
- Ei block ta run korle kintu kichu print hobe na! Amra shudhu function ta *define* korlam.

## 2. Function ke daka (Calling)

Function use korar jonno amader setake "call" korte hoy.

```python
say_hello()
say_hello()
```

### Bakhya (Explanation)
- Amra function er nam likhi ebong tarpor bracket `()` dei.
- Amra duibar call korechi tai eta duibar print korbe.

## 3. Arguments (Tottho pathano)

Amra function er bhitore tottho (data) pathate pari jate kore function ti alada alada kaj korte pare.

```python
def greet_person(name):
    print("Hello, %s!" % name)
```

```python
greet_person("Alice")
greet_person("Bob")
```

### Bakhya (Explanation)
- Amra definition er somoy bracket er bhitore `name` likhechi.
- Jokhon amra `greet_person("Alice")` call kori, tokhon `"Alice"` value ti function er bhitore `name` variable e chole jay.

## 4. Return Values (Folafol ferot deya)

Function prayi kichu hiseb-nikesh kore ebong folafol amader ferot dey. Er jonno amra `return` use kori.

```python
def add_numbers(x, y):
    result = x + y
    return result
```

```python
suma = add_numbers(5, 3)
print("Jogfol holo: %d" % suma)
```

### Bakhya (Explanation)
- `return result` value tike ferot pathay jekhan theke function call kora hoyechilo.
- `suma = add_numbers(5, 3)` er mane `suma` variable er man hobe 8.

### Problem D-1: Calculate Factorial

**Question (প্রশ্ন):** 
Write a function to calculate the factorial value of any integer entered through the keyboard.

**Explanation (ব্যাখ্যা):**
- Factorial (n!) = n × (n-1) × (n-2) × ... × 1
- উদাহরণ: 5! = 5 × 4 × 3 × 2 × 1 = 120
- 0! = 1 (বিশেষ ক্ষেত্র)

```python
# Solution D-1: Calculate Factorial
def factorial(n):
    """
    Function to calculate factorial of a number
    বাংলা: একটি সংখ্যার ফ্যাক্টোরিয়াল গণনা করার ফাংশন
    """
    if n < 0:
        return "Factorial not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Test the function
print("=== Factorial নির্ণয় করা ===")
num = int(input("Enter a number: "))
fact = factorial(num)
print(f"{num}! = {fact}")

# Additional test cases
print("\nTest cases:")
for i in range(6):
    print(f"{i}! = {factorial(i)}")
```

### Problem D-2: Power Function

**Question (প্রশ্ন):**
Write a function power(a,b) to calculate the value of a raised to b.

**Explanation (ব্যাখ্যা):**
- a^b = a × a × a × ... (b times)
- উদাহরণ: 2^5 = 2 × 2 × 2 × 2 × 2 = 32
- বিশেষ ক্ষেত্র: যেকোনো সংখ্যার ঘাত 0 হলে ফলাফল 1

```python
# Solution D-2: Power Function
def power(a, b):
    """
    Calculate a raised to the power b (a^b)
    বাংলা: a কে b এর ঘাত দিয়ে গুণ করা
    """
    result = 1
    for i in range(b):
        result *= a
    return result

# Test the function
print("=== Power (ঘাত) গণনা করা ===")
base = int(input("Enter base (ভিত্তি): "))
exponent = int(input("Enter exponent (ঘাত): "))
result = power(base, exponent)
print(f"{base}^{exponent} = {result}")

# Alternative using ** operator
print(f"Using ** operator: {base} ** {exponent} = {base**exponent}")

# Test cases
print("\nTest cases:")
test_cases = [(2, 5), (3, 3), (5, 2), (10, 3), (2, 0)]
for base, exp in test_cases:
    print(f"{base}^{exp} = {power(base, exp)}")
```

### Problem D-3: Calculate LCM

**Question (প্রশ্ন):**
Write a function to calculate LCM of two numbers.

**Explanation (ব্যাখ্যা):**
- LCM (Least Common Multiple) = সর্বনিম্ন সাধারণ গুণিতক
- LCM গণনা করার পদ্ধতি: (a × b) / GCD(a, b)
- উদাহরণ: LCM(12, 18) = 36

```python
# Solution D-3: Calculate LCM
def gcd(a, b):
    """
    Calculate GCD (Greatest Common Divisor) using Euclidean algorithm
    বাংলা: ইউক্লিডীয় অ্যালগরিদম ব্যবহার করে GCD বের করা
    """
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    """
    Calculate LCM (Least Common Multiple)
    বাংলা: সর্বনিম্ন সাধারণ গুণিতক বের করা
    Formula: LCM(a, b) = (a * b) / GCD(a, b)
    """
    return (a * b) // gcd(a, b)

# Test the function
print("=== LCM গণনা করা ===")
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
gcd_result = gcd(num1, num2)
lcm_result = lcm(num1, num2)
print(f"GCD({num1}, {num2}) = {gcd_result}")
print(f"LCM({num1}, {num2}) = {lcm_result}")

# Test cases
print("\nTest cases:")
test_cases = [(12, 18), (15, 25), (10, 20), (7, 13)]
for a, b in test_cases:
    print(f"LCM({a}, {b}) = {lcm(a, b)}, GCD({a}, {b}) = {gcd(a, b)}")
```

### Problem D-4: Calculate GCD

**Question (প্রশ্ন):**
Write a function to calculate GCD of two numbers.

**Explanation (ব্যাখ্যা):**
- GCD (Greatest Common Divisor) = সর্বোচ্চ সাধারণ ভাজক
- ইউক্লিডীয় অ্যালগরিদম ব্যবহার করে দ্রুত গণনা করা যায়
- উদাহরণ: GCD(12, 18) = 6

```python
# Solution D-4: Calculate GCD (GCD already defined above, here's additional implementation)
def gcd_simple(a, b):
    """
    Calculate GCD using Euclidean algorithm (iterative approach)
    বাংলা: ইউক্লিডীয় অ্যালগরিদম ব্যবহার করে GCD বের করা
    """
    while b != 0:
        temp = b
        b = a % b
        a = temp
    return a

# Test the function
print("=== GCD গণনা করা ===")
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
gcd_result = gcd_simple(num1, num2)
print(f"GCD({num1}, {num2}) = {gcd_result}")

# Test cases
print("\nTest cases:")
test_cases = [(12, 18), (48, 18), (100, 50), (17, 19), (21, 14)]
for a, b in test_cases:
    print(f"GCD({a}, {b}) = {gcd_simple(a, b)}")
```

### Problem D-5: Leap Year Checker

**Question (প্রশ্ন):**
Any year is entered through the keyboard. Write a function to determine whether the year is a leap year or not.

**Explanation (ব্যাখ্যা):**
- Leap year শর্ত:
  - বছর 400 দ্বারা বিভাজ্য হলে leap year
  - অথবা বছর 4 দ্বারা বিভাজ্য কিন্তু 100 দ্বারা নয় হলে leap year
  - অন্যথায় leap year নয়
- উদাহরণ: 2020 leap year, 2021 নয়, 2000 leap year

```python
# Solution D-5: Leap Year Checker
def is_leap_year(year):
    """
    Determine if a year is a leap year
    বাংলা: কোনো বছর leap year কিনা তা নির্ণয় করা
    
    Rules:
    - Divisible by 400 => Leap year
    - Divisible by 100 (but not 400) => Not leap year
    - Divisible by 4 (but not 100) => Leap year
    - Otherwise => Not leap year
    """
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

# Test the function
print("=== Leap Year নির্ণয় করা ===")
year = int(input("Enter a year: "))
if is_leap_year(year):
    print(f"{year} is a leap year (leap year)")
else:
    print(f"{year} is not a leap year (leap year নয়)")

# Test cases
print("\nTest cases:")
test_years = [2000, 2001, 2004, 2100, 2020, 2021, 1900, 2024]
for y in test_years:
    result = "Leap" if is_leap_year(y) else "Not Leap"
    print(f"{y}: {result}")
```

### Problem D-6: Prime Factors

**Question (প্রশ্ন):**
An integer is entered through the keyboard. Write a function to obtain the prime factors of this number. For example, prime factors of 24 are 2, 2, 2, and 3, whereas prime factors of 35 are 5 and 7.

**Explanation (ব্যাখ্যা):**
- Prime factors বের করার পদ্ধতি:
  - 2 দিয়ে যতক্ষণ সম্ভব ভাগ করুন
  - তারপর 3, 5, 7, ... বিজোড় সংখ্যা দিয়ে ভাগ করুন
  - শেষে যা থাকে তা প্রাইম ফ্যাক্টর
- উদাহরণ: 24 = 2 × 2 × 2 × 3

```python
# Solution D-6: Prime Factors
def prime_factors(n):
    """
    Get all prime factors of a number
    বাংলা: একটি সংখ্যার সমস্ত প্রাইম ফ্যাক্টর বের করা
    """
    factors = []
    
    # Divide by 2 until it's odd
    while n % 2 == 0:
        factors.append(2)
        n = n // 2
    
    # Divide by odd numbers starting from 3
    i = 3
    while i * i <= n:
        while n % i == 0:
            factors.append(i)
            n = n // i
        i += 2
    
    # If n is still greater than 1, it's a prime factor
    if n > 1:
        factors.append(n)
    
    return factors

# Test the function
print("=== Prime Factors বের করা ===")
num = int(input("Enter a number: "))
factors = prime_factors(num)
print(f"Prime factors of {num}: {factors}")
print(f"Product representation: {' × '.join(map(str, factors))}")

# Test cases
print("\nTest cases:")
test_numbers = [24, 35, 100, 17, 90, 120]
for n in test_numbers:
    factors = prime_factors(n)
    print(f"Prime factors of {n}: {factors}")
```
