## Set A
1. Write a program to print the following pattern (Multiplication Table). The number of rows and columns will be the same and must be taken as user input.


```python
n = int(input('Enter the number of rows/columns: '))
for i in range(1, n + 1):
    for j in range(1, n + 1):
        print(i * j, end='	')
    print()
```

2. Write a program using a loop to count how many even numbers are between 1-25.


```python
count = 0
for i in range(1, 26):
    if i % 2 == 0:
        count += 1
print(f'Total even numbers between 1-25: {count}')
```

## Set B
1. Write a program to print the following pattern (Reverse Descending Triangle). The number of rows must be taken as user input.


```python
n = int(input('Enter the number of rows: '))
for i in range(n, 0, -1):
    for j in range(i, 0, -1):
        print(j, end=' ')
    print()
```

2. Write a program using a loop to print all numbers less than 50 that are divisible by 4 or 6.


```python
print('Numbers less than 50 divisible by 4 or 6:')
for i in range(1, 50):
    if i % 4 == 0 or i % 6 == 0:
        print(i, end=' ')
```

## Set C
1. Write a program to print the following pattern (Hollow Square). The number of rows and columns will be the same and must be taken as user input.


```python
n = int(input('Enter the number of rows/columns: '))
for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
```

2. Write a program using a loop to print all numbers less than 20 that are divisible by both 3 and 5.


```python
print('Numbers less than 20 divisible by both 3 and 5:')
for i in range(1, 20):
    if i % 3 == 0 and i % 5 == 0:
        print(i)
```

## Set D
1. Write a program to print the following pattern (Diagonal Number Pattern). The number of rows and columns will be the same and must be taken as user input.


```python
n = int(input('Enter the number of rows/columns: '))
for i in range(1, n + 1):
    for j in range(1, n + 1):
        if i == j:
            print(i, end=' ')
        else:
            print(0, end=' ')
    print()
```

2. Write a program using a loop to find the sum of all odd numbers between 1-50.


```python
odd_sum = 0
for i in range(1, 51):
    if i % 2 != 0:
        odd_sum += i
print(f'The sum of odd numbers between 1-50 is: {odd_sum}')
```
