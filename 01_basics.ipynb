## 2. Variables

Variables holo baksho (box) er moto jekhane amra data stor kore rakhte pari. Amra suvidhar jonno egulor ekta nam dei.

```python
my_number = 10
my_name = "Student"

print(my_number)
print(my_name)
```

### Bakhya (Explanation)
- `my_number = 10`: Amra `my_number` name ekta baksho banalam ebong tar bhitore `10` rakhlam.
- `my_name = "Student"`: Amra `my_name` name arekta baksho banalam ebong tar bhitore `"Student"` text ti rakhlam.
- Tarpor `print` use kore amra oigular bhitore ki ache ta dekhlam.

## 3. Comments

Comments holo emon kichu line ja Python ignore kore (run kore na). Egulo shudhu manush der porar suvidhar jonno.

```python
# Eta ekta comment. Python ei line ta porbe na.
print("This will be printed.") # Amnra line er shesheo comment likhte pari.
```

## 4. Basic Math

Python calculator er moto onko (math) korte pare.

```python
a = 5
b = 2

sum_result = a + b      # Jog (Addition)
diff_result = a - b     # Biyog (Subtraction)
prod_result = a * b     # Gun (Multiplication)
div_result = a / b      # Bhag (Division)

print(sum_result)
print(diff_result)
print(prod_result)
print(div_result)
```

### Bakhya (Explanation)
- `+` diye jog kora hoy.
- `-` diye biyog kora hoy.
- `*` diye gun kora hoy.
- `/` diye bhag kora hoy.

## 5. String Formatting (Classic Style)

Majhe majhe amra text ar variable eksathe mix kore dekhate chai. Er jonno `%` operator use kora hoy. Eta onek purono kintu kajer poddhoti.

```python
age = 25
name = "Alice"

print("Amar nam holo %s ebong amar boyosh %d bochor." % (name, age))
```

### Bakhya (Explanation)
- String ba text er bhitore, `%s` use kora hoy String er jonno.
- `%d` use kora hoy Number ba Decimal er jonno.
- String er pore, `% (name, age)` deya hoy.
- Python `name` er value ta `%s` er jagay, ar `age` er value ta `%d` er jagay bosiye dey.

```python
length=float(input("Enter the length of the rectangle: "))
width=float(input("Enter the width of the rectangle: "))
radius=float(input("Enter the radius of the circle: "))

area_rectangle=length*width
perimeter_rectangle=2*(length+width)
area_circle=3.1416*radius*radius
circumference_circle=2*3.1416*radius
print(f"Area of Rectangle: {area_rectangle}")
print(f"Perimeter of Rectangle: {perimeter_rectangle}")
print(f"Area of Circle: {area_circle}")
print(f"Circumference of Circle: {circumference_circle}")
```
